````python

def AMi_candidateline(line):
    """
      input:   line 
      output:  True/False
      decide whether line is candidate by checking 
      1. slope of the line > minimum_slope
      2. (slope_of_line - avg_slope_line) < max_slope_difference
      3. y2_of_line should be below the vanishing point of lane (right and left line meets in the middle on the top of image)
      4. line should not be far from lane line 
    """
    for x1,y1,x2,y2 in line:
        if (x2-x1) == 0: 
            print("inifinte_slope")
            return False    #infinite slope not candidate line
        m_slope = (y2-y1)/(x2-x1)
        if (abs(m_slope) < LaneParameters.HORIZONTAL_OR_MIN_SLOPE):  # less than minimum SLOPE or horizantal line not candidate
            print("horizontal_slope")
            return False
        if (m_slope > 0):  #right line
            right_m_slope = getattr(Lane,'right_m_avg')
            if (abs(m_slope) - right_m_slope) > LaneParameters.MAX_SLOPE_DIFFERENCE:  # above threshold value of average slope not a candidate
                print("max_slope_diff")
                return False
            distance_from_line =  distance_to_lane_line('right_line',x1,y1,x2,y2)
            if (distance_from_line is not None):
                if (distance_from_line > LaneParameters.MAX_DISTANCE_FROM_LINE):
                    print("distance from line max")
                    return False   
            if (Lane.left_line is not None):   #if not first frame
                if (y2 < Lane.left_line.threshold_point[1]):  #point at which both left and right lines merge is threshold; cross threshold not a candidate
                    print("beyond thresholdpoint")
                    return False
        else:
            left_m_slope = getattr(Lane,'left_m_avg')
            if (abs(m_slope) - left_m_slope) > LaneParameters.MAX_SLOPE_DIFFERENCE:  #not candidate
                print("max_slope_diff")
                return False
            distance_from_line =  distance_to_lane_line('left_line',x1,y1,x2,y2)
            if (distance_from_line is not None):
                if (distance_from_line > LaneParameters.MAX_DISTANCE_FROM_LINE):
                    print("max_distance_from_line")
                    return False
            if (Lane.left_line is not None):
                if (y2 < Lane.left_line.threshold_point[1]):
                    print("beyond threshold point")
                    return False
    print("returning true")   
    return True  

def distance_to_lane_line(line_dir,x1,y1,x2,y2):
        """
        Reference https://en.wikipedia.org/wiki/Distance_from_a_point_to_a_line
        """
        lane_line = getattr(Lane, line_dir)
        if lane_line is None: return None
        avg_x = (x2 + x1) / 2
        avg_y = (y2 + y1) / 2
        distance = abs(lane_line.coeffs[0] * avg_x - avg_y + lane_line.coeffs[0])/math.sqrt(lane_line.coeffs[0] ** 2+1)
        return distance

class LaneParameters(object):
    
    SLOPE_TOLERANCE = 0.1
    HORIZONTAL_OR_MIN_SLOPE = 0.4
    MAX_SLOPE_DIFFERENCE = 0.8
    MAX_DISTANCE_FROM_LINE = 20
  
    visual_offset = 20

    #parameters of pipeline
    #blur kernel size
    kernel_size = 3
    #threshold(min,max) of canny algorithm
    low_threshold = 50;
    high_threshold = 150;
    #region of interest vertices in terms of iamge sizes
    region_base = 0.85  #80% of imgae.shape[0]
    region_top = 0.07  #top edge of region in terms image size
    region_height = 0.4  #top edge of region  in terms of image size

    #hough line parameters
    rho=1  # distance resoultion in pixels
    theta = np.pi/180  # angualar resolution in radian f the grid
    #It is difficult to detect straight lines which are part of a curve because they are very very small. 
    #For detecting such lines it is important to properly set all the parameters of Hough transform.
    #Two of most important parameters are: Hough votes and maximum distance between points 
    #which are to be joined to make a line. Both parameters are set at their minimum value.
    threshold = 15   
    min_line_length = 10
    color_thickness = 5
    max_line_gap=20
    
    img_shape_0=960
    img_shape_1=480
    
    FIRST_FRAME_LINE_RANGES = {'left_line' : range(480),
                               'right_line' : range(480,960)}
            
    def __init__(self,min_slope,slop_diff,distance_from_line,slope_tolerance):
        self.SLOPE_TOLERANCE = slope_tolerance
        self.MAX_SLOPE_DIFFERENCE = slope_diff
        self.HORIZONTAL_OR_MIN_SLOPE = min_slope
        self.MAX_DISTANCE_FROM_LINE = distance_from_line
        

class Lane(object):
    
    COLORS = {
        'lane_color' : (255,0,0),
        'region_stable' : (0,80,60),
        'region_unstable' : (255,80,60),
        'left_line' : (60,40,220),
        'right_line' : (255,0,255)
    }
    
    THICKNESS = 5
    
    left_line = None
    right_line = None
    left_m_avg =0.5
    right_m_avg = 0.5
    
    
    @staticmethod
    def lines_exist():
        return all([Lane.left_line,Lane.right_line])
    
    def compute_slope(line):
        line_x1,line_y1,line_x2,line_y2 = line[0]
        slope = (line_y2 - line_y1)/(line_x2 - line_x1)
        return slope
    def compute_intercept(line):
        line_x1,line_y1,line_x2,line_y2 = line[0]
        intercept = line_y1 - Lane.compute_slope(line)*line_x1
        return intercept
    
    def get_y(self,slope,intercept,x):
        return (slope*x+intercept)
    
    @staticmethod
    def fit_lane_line(lines):
        """
        polyfit line segments and generate efficients   
        separate out x,y points from segments 
        if first frame, apply different selection points
     
        """
        x,y = [], []
      
        for line in lines:
            if (AMi_candidateline(line)):
                line_x1,line_y1,line_x2,line_y2 = line[0]
                x_coords = list(range(line_x1, line_x2, 1))
                y_coords = list(map(lambda x: Lane.get_y(Lane.compute_slope(line),Lane.compute_intercept(line),x), x_coords))
                x.extend(x_coords)
                y.extend(y_coords)
        
        #handle first frame differently
        if x != [] and not Lane.lines_exist():
            line  = lines[0]
            if (compute_slope(line) > 0):
                coords = np.array([[x,y] for x,y in zip(x,y) if x in LaneParameters.FIRST_FRAME_LINE_RANGES['right_line']])
                x = coords[:,0] 
                y = coords[:,1]
            else:
                coords = np.array([[x,y] for x,y in zip(x,y) if x in LaneParameters.FIRST_FRAME_LINE_RANGES['left_line']])
                x = coords[:,0]
                y = coords[:,1]
                
        if x != []:
            coeffs = np.polyfit(x,y,1)
            return coeffs,list(zip(x,y)) 
        else:
            return None, None
     
      
    @staticmethod
    def update_threshold_point(left,right):
        coeffs_diff = left.coeffs - right.coeffs
        x = -coeffs_diff[1] / coeffs_diff[0]
        y = np.poly1d(left.coeffs)(x)
        x, y = map(int, [x,y])
        left.threshold_point = [x,y]
        right.threshold_point = [x,y]
     
    @staticmethod
    def delete():
        Lane.left_line = None
        Lane.right_line = None    
        Lane.left_m_avg = 0.5
        Lane.right_m_avg = 0.5
                
    def __init__(self, segments):
                      
        self.coeffs_array = np.array([])        
        # Lane coefficients from the current image
        print("calling fit_lane_line")
        self.current_lane_line_coeffs, self.points = Lane.fit_lane_line(segments)
           
        
        if self.current_lane_line_coeffs is None: 
            raise Exception('Cannot initialize lane. No lines detected.')
        
        # array of coeffs for tracking average slope,intercepts over frames
        self.coeffs_array = np.append(self.coeffs_array,[self.current_lane_line_coeffs],axis=0)
        
        # coefficients of lane line
        self.coeffs = self.coeffs_array[0]
        
        
        # Stability flag. Set to False if the slope changes too rapidly
        self.stable = True
        
        # Hough lines
        self.hough_lines = None
        
        # List of points which belong to this lane line. Transformed from segments
        self.points = None
        
        # Coordinates for drawing this lane line
        self.x1, self.x2, self.y1, self.y2 = 0,0,0,0
                
    @property
    def a(self):
        """
        Slope of the lane line. 
        """
        if len(self.coeffs) == 2: 
            return self.coeffs[0]
        else: 
            return 1   
    
    @property
    def b(self):
        """
        Intercept of the lane line. 
        """
        if len(self.coeffs) == 2: 
            return self.coeffs[1]        
        else:
            return 1    
    def update_lane_line(self, lines):
        """
             update average slope if current frame slope is within tolerance level from average slope
             use average slope to calculate line parameters x1,y1,x2,y2
             
        """
        mean_coeffs = np.mean(self.coeffs_array,axis=0)
                
        self.coeffs = np.mean(self.coeffs_array,axis=0)
        #update avereage m values 
        if (self.coeffs[0] > 0):  #right lane line
            self.right_m_avg = self.coeffs[0]
        else:    
            self.left_m_avg  = self.coeffs[0]
        current_frame_lane_line_coeffs, current_frame_line_points = Lane.fit_lane_line(lines)  
        if current_framt_lane_line_coeffs is None:
            current_frame_lane_line_coeffs = mean_coeffs
        if current_frame_line_points is not None:
            self.points = current_frame_line_points
        m_mean = mean_coeffs[0]
        m_current = current_frame_lane_line_coeffs[0]
        if abs(m_mean-m_current) > LaneParameters.SLOPE_TOLERANCE:
            self.coeffs_array = np.append(self.coeffs_array,current_frame_lane_line_coeffs,axis=0)
        else:
            self.coeffs_array = np.append(self.coeffs_array,(np.dot([.1,.9],np.vstack[current_frame_lane_line_coeffs,mean_coeffs])),axis=0)
        self.update_lane_line_coords()
                
                
    def update_segments_list(self, lines):
        self.hough_lines = lines
    
    def get_x_coord(self, y):
        return int((y - self.coeffs[1]) / self.coeffs[0])
    
    def update_lane_line_coords(self):
        # Offset to distinguish lines
        self.y1 = LaneParameters.img_shape_1
        self.x1 = self.get_x_coord(self.y1)
        self.y2 = self.threshold_point[1] + LaneParameters.visual_offset   #offset to distinguish lines
        self.x2 = self.get_x_coord(self.y2)         
                
````

````python
import math

def regionAction(img):
    imshape = img.shape
    #vertices = np.array([[(0,imshape[0]),(((imshape[1]/2)-20),(imshape[0]/2)+40),((imshape[1]/2+20),(imshape[0]/2)+40),(imshape[1],imshape[0])]],dtype=np.int32)
    #vertices = np.array([[(0,imshape[0]),(((imshape[1]/2)),((imshape[0]/2)+10)),(imshape[1],imshape[0])]],dtype=np.int32)
    vertices = np.array([[((imshape[1]*(1-LaneParameters.region_base))//2,imshape[0]),\
                          ((imshape[1]*(1-LaneParameters.region_top))//2, (imshape[0] - (imshape[0]*LaneParameters.region_height))),\
                          (imshape[1]-(imshape[1]*(1-LaneParameters.region_top))//2, (imshape[0]-(imshape[0]*LaneParameters.region_height))),\
                          (imshape[1]-(imshape[1]*(1-LaneParameters.region_base))//2, imshape[0])]],dtype=np.int32)  
                    
    #vertices = np.array([[(30,imshape[0]),(450, 315), (500, 315), (imshape[1],imshape[0])]], dtype=np.int32)                    
    return(region_of_interest(img,vertices))

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    #return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def intersection_points(left_coeffs,right_coeffs):
    equation = left_coeffs.coeffs - right_coeffs.coeffs
    x = - equation[1] / equation[0]
    y = np.poly1d(left_coeffs.coeffs)(x)
    x, y = map(int, [x,y])
    intersection_points = [x,y]
    return intersection_points[1]

def draw_eachline(img,x,y,color=[255,0,0], thickness=10):
    """
    Draw line to the points[x,y] using color =[255,0,0] and thickness
    """
    #empty array
    if len(x) == 0:
        return
    
    line_coefficients = np.polyfit(x,y,1)   #1 (line equation)
    m_slope = line_coefficients[0]  #do we get average of all line segment
    b_recp  = line_coefficients[1]
    
    ysize = img.shape[0]
    xsize = img.shape[1]
    
    y1 = ysize;
    x1 = int((y1-b_recp)/m_slope)  # y = mx +b x = y-b/m
    y2 = int(ysize/2)+50  # find another point(y) in image shape middle + some base
    x2 = int((y2-b_recp)/m_slope)
    
    cv2.line(img,(x1,y1),(x2,y2),[255,0,0],4)

def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # array to collect line segments of left &right side
    left_segmentpointsX = []
    left_segmentpointsY = []
    right_segmentpointsX = []
    right_segmentpointsY = [] 
    
    # m =  y2-y1/x2-x1 
    # m > 0 left line segments
    # m < 0 right line segments
    
    #slope tolerance level
    m_threshold = 0.5
    left_draw = True
    right_draw = True
    
    for line in lines:
        x1,y1,x2,y2 = line[0]
        if (x2-x1) == 0:
            m_slope =  999    #infinite slope
        else:
            m_slope = (y2-y1)/(x2-x1)
        if  (abs(m_slope)  >   m_threshold):  # filter out horizontal lines  and resonaable slopes (and lines )
                if  m_slope > 0:    #right lane
                    right_segmentpointsX.append(x1)
                    right_segmentpointsY.append(y1)
                    right_segmentpointsX.append(x2)
                    right_segmentpointsY.append(y2)                 
                else:  
                    left_segmentpointsX.append(x1)
                    left_segmentpointsY.append(y1)
                    left_segmentpointsX.append(x2)
                    left_segmentpointsY.append(y2)   
   

    if (len(left_segmentpointsX)):
        m_left,b_left  = np.polyfit(left_segmentpointsX,left_segmentpointsY,1)
    else:
        left_draw = False
        m_left, b_left = 1,1
    if (len(right_segmentpointsX)):    
        m_right,b_right  = np.polyfit(right_segmentpointsX,right_segmentpointsY,1)
    else:
        right_draw = False
        m_right,b_right = 1,1
    

    ymax = img.shape[0]
    xmax = img.shape[1]
    
    y1  = ymax
    y2  = ymax*(1-LaneParameters.region_height)
    
    
    #left_x2 = (y2-b_avg_left)/m_avg_left
    #left_x1 = (y1-b_avg_left)/m_avg_left
    #right_x2 = (y2-b_avg_right)/m_avg_right
    #right_x1  = (y1-b_avg_right)/m_avg_right
  
    left_x2   = (y2-b_left)/m_left
    left_x1   = (y1-b_left)/m_left
    right_x2  = (y2-b_right)/m_right
    right_x1  = (y1-b_right)/m_right

    if (left_draw):
        left_x1 = int(left_x1)
        left_x2 = int(left_x2)
        y2 = int(y2)
        y1 = int(y1)
        cv2.line(img,(left_x1,y1),(left_x2,y2),color,thickness)
    if (right_draw):
        right_x1 = int(right_x1)
        right_x2 = int(right_x2)
        y2 = int(y2)
        y1 = int(y1)
        cv2.line(img,(right_x1,y1),(right_x2,y2),color,thickness)
        
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    candidate_lines = []
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    if lines is not None:
        candidate_lines = select_candidatelines(lines)
        print(len(candidate_lines))
        return candidate_lines
    else:
        return  None
    
    #line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    #draw_lines(line_img, lines)
    #return line_img

def hough_lines_and_draw(img,rho, theta, threshold, min_line_len, max_line_gap):
    #lines = cv2.HoughLinesP(img, rho, theta, threshold, min_line_len, max_line_gap)
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    if lines is not None:
        line_img = np.zeros((img.shape[0],img.shape[1],3), dtype=np.uint8)
        draw_lines(line_img,lines)
        return line_img
    else:
        return None
          

def select_candidatelines(lines):
    """
      lines are the output of Hough_transform() 
      this function selects candidates from lines using selection algorithm AMi_candidate
      returns left and right candidate lines
    """
    
    candidate_lines = []
    for line in lines:
        if (AMi_candidateline(line)): 
               candidate_lines.append(line)               
    return   candidate_lines
 
    
# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )

def  saveImages(images, outputDir, imageNames, isGray=0):
    
    """
       saves images in outputdir + 'img' using iamgeNames
       
       Example : sameImages([img], 'testDir', ['gray_images'])
       will save image as testDir/gray_images
    """
    if not os.path.isdir(outputDir):
            os.mkdir(outputDir)
            
    list_imgs = list(map(lambda imgList: (outputDir+imgList[1], imgList[0]), zip(images,imageNames)))
    
    for imgPair in list_imgs:
        if isGray:
            plt.imsave(imgPair[0],imgPair[1],cmap='gray')
        else :
            plt.imsave(imgPair[0],imgPair[1])
            
def readandDisplay(inputDir,ImageNames, isGray=0):
    """
      read each images from inputdir and show using imageplot
      if gray show gray images
      
      example('test_images',['img0'],0)
    
    """
    DisImages = list(map(lambda img: mpimg.imread(inputDir +'/' + img),ImageNames))
    for img in DisImages:
    #gray_img = grayscale(img)
        if isGray:
            plt.imshow(img,cmap='gray')
        else:
            plt.imshow(img)
        plt.show()
        imshape = img.shape
        #print(imshape[0])
        #print(imshape[1])

def get_slope(line):
    for x1,y1,x2,y2 in line:
        return ((y2-y1)/(x2-x1));
                
def update_lane(line_segments):
    if line_segments is not None:
        left = [segment for segment in line_segments if get_slope(segment) < 0]
        right = [segment for segment in line_segments if get_slope(segment) > 0]
        print("update_lane:")
        print(len(left))
        print(len(right))
        if not Lane.lines_exist():
            Lane.left_line = Lane(left)
            Lane.right_line = Lane(right)      
        Lane.update_threshold_point(Lane.left_line, Lane.right_line)
        Lane.left_line.update_lane_line([l for l in left if AMi_candidateline(l)])
        Lane.right_line.update_lane_line([r for r in right if AMi_candidateline(r)])      
        
def lanefinding_pipeline(img,FirstFrame=0):
       
    #stage 1 convert image to grayscale
    gray_img = grayscale(image)
    #stage 2 apply gaussian blur 
    blur_img = gaussian_blur(gray_img,LaneParameters.kernel_size)
    #stage 3 canny edge detection
    edge_img  = canny(blur_img,LaneParameters.low_threshold,LaneParameters.high_threshold)
    #stage 4 region of Interest
    region_img = regionAction(edge_img)
    #stage 5 hough lines
    line_segments = hough_lines(region_img,LaneParameters.rho,LaneParameters.theta,LaneParameters.threshold,LaneParameters.min_line_length,LaneParameters.max_line_gap)
    
    #stage 6  draw lane
    update_lane(line_segments)
    
    #stage 7 
    line_img = np.zeros_like(img)
    for line in [Lane.left_line, Lane.right_line]:
        line.update_lane_line_coords()
        cv2.line(line_img, (line.x1, line.y1), (line.x2, line.y2), [255, 0, 0], LaneParameters.THICKNESS)  
    final_img = weighted_img(line_img,img)                        
    return final_img
        
````
